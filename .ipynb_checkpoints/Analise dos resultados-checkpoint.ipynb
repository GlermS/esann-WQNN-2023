```python
results = []
for key in dfs:
    transp = []
    MAE = []
#     for dfx in dfs[str(key)].values:
#         df = dfx.copy().query('ghi > 0')
    df= dfs[key].copy().query('ghi > 0')
    transp.append(df.POA.sum()/df.ghi.sum())
    MAE.append((df.angle-df.mangle).abs().mean())
#     lista= key.split('_')
    cfg = {}
    cfg.update({
        'transposition_gain': np.mean(transp),
        'MAE': np.mean(MAE)
    })
    
#     for item  in lista:
#         k, n = item.split('=')
#         cfg[k] = float(n)
    cfg['index'] = int(key.replace('checkpoint_','').replace('.csv',''))
    results.append(cfg)
```
