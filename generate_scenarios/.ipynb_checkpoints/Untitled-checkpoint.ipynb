```python
import sys
sys.path.append('..')
from utils import solar, plotUtils
import pvlib
import pandas as pd

from scipy import stats
```

```python
plant_info = { # Macaé
        "latitude": -22.3717,
        "longitude": -41.7857,
        "altitude": 0,
        "azimuth": 90,
        "gcr": 0,
        'tz':0,
        'albedo':0.2
    }

startDate, endDate = '2022-01-01 00:00', '2022-12-31 23:55'
```

```python
dfCs = solar.getClearSkyModelDf(
    plant_info
    , startDate, endDate, freq='15min'
    )
```

```python
def generate_scenario(dfx, shape_factor= 2, scale_factor=1/100):
    df = dfx.copy()
    
    dist = stats.gamma(shape_factor, scale=scale_factor)
    x = dist.rvs(len(df))
    n = [min(max(0,i), 1) for i in x]
    df['kc'] = n
    
    df['ghi'] = df['mghi']*(1-df['kc'])
    
    a = pvlib.irradiance.erbs(
        ghi=df['ghi'].values,  
        zenith=df['zenith'].values*-1, 
        datetime_or_doy=dfCs.index
    )

    df['dni'] = a['dni']
    df['dhi'] = a['dhi']
    df['kt'] = a['kt']
    
    
    dfg = pvlib.irradiance.get_total_irradiance(
    surface_tilt=-df['mangle'],
    surface_azimuth=plant_info['azimuth'],
    dni=df['dni'],
    ghi=df['ghi'],
    dhi=df['dhi'],
    solar_zenith=-df['zenith'],
    solar_azimuth=df['azimuth'], albedo=plant_info['albedo']
    )
    
    df['mg'] = dfg['poa_global']   
    return df
    
```

```python
df = pd.read_csv('../data/ws1-macae.csv', sep=';',decimal=',')
index = []
for date, hour in df[['Data','Hora (UTC)']].values:
    index.append(pd.Timestamp(date).replace(hour=int(hour/100)))
    
df.index = index

dfx = dfCs.assign(ghi=df.resample('15min').mean()['Radiacao (KJ/m²)'])
dfx.loc[dfx.mg == 0, 'ghi'] = 0
dfx['ghi'] = dfx['ghi'].interpolate()/3.28
```

```python
df = dfx.copy()
a = pvlib.irradiance.erbs(
    ghi=df['ghi'].values,  
    zenith=df['zenith'].values*-1, 
    datetime_or_doy=dfCs.index
)
```

```python
df['dni'] = a['dni']
df['dhi'] = a['dhi']
```

```python
dfg = pvlib.irradiance.get_total_irradiance(
    surface_tilt=-df['mangle'],
    surface_azimuth=plant_info['azimuth'],
    dni=df['dni'],
    ghi=df['ghi'],
    dhi=df['dhi'],
    solar_zenith=-df['zenith'],
    solar_azimuth=df['azimuth'], albedo=plant_info['albedo']
    )
    
df['mg'] = dfg['poa_global']
```

```python
import os
class LoadDfs:
    path
    def __iter__(self):
        self.path = path
        self.files = os.listdir(self.path)
        self.i = 0
        return self
    def __next__(self):
        if self.i >= len(self.files):
            raise StopIteration
            
        file = self.files[self.i]
        df = pd.read_csv(os.path.join(self.path, file), index_col='ts', parse_dates=['ts'])
        self.i+= 1            
        return file, df
```
